# 3. Con los archivos fallecidosCOVID.csv,  reporte mensual de fallecidos. Mostrar mes (y año) y cuantos fallecidos hubo.

```python
# Leer el archivo fallecidosCOVID.csv
fallecidos = pd.read_csv("https://github.com/in-Sinergy/Pau_T3/raw/main/fallecidosCOVID.csv")
fallecidos
```

```python
#se observa que las columnas están todas juntas
#hay que separarlas
fallecidos[['corte','fallecimiento', 'edad', 'sexo', 'criterio', 'dep', 'prov', 'dist', 'ubigeo', 'uuid']] =fallecidos['FECHA_CORTE;FECHA_FALLECIMIENTO;EDAD_DECLARADA;SEXO;CLASIFICACION_DEF;DEPARTAMENTO;PROVINCIA;DISTRITO;UBIGEO;UUID'].str.split(';', expand=True)
fallecidos
```

```python
#eliminar la primera columna
fallecidos = fallecidos.drop(columns='FECHA_CORTE;FECHA_FALLECIMIENTO;EDAD_DECLARADA;SEXO;CLASIFICACION_DEF;DEPARTAMENTO;PROVINCIA;DISTRITO;UBIGEO;UUID')
fallecidos
```

```python
fallecidos
```

```python
#eliminar las columnas 'corte' y 'uuid' ya que son innecesarias
bye=['corte', 'uuid']
fallecidos = fallecidos.drop(columns=bye)
fallecidos
```

```python
#comienza el formateo de las columnas
# Convertir la columna de fecha a datetime
pd.to_datetime(fallecidos['fallecimiento'])
fallecidos['fecha'] = pd.to_datetime(fallecidos['fallecimiento'])
```

```python
# Extraer el año y el mes de la fecha
fallecidos['año'] = fallecidos['fecha'].dt.year
fallecidos['mes'] = fallecidos['fecha'].dt.month
```

```python
# Agrupar por año y mes, y contar los fallecidos
reporte_mensual = fallecidos.groupby(['año', 'mes']).size().reset_index(name='fallecidos')
```

```python
# Mostrar el reporte mensual
reporte_mensual
```

# 4. Reporte de fallecidos por mes, pero el tipo de muerte debe estar en columnas separadas.

```python
# Agrupar por año, mes y tipo de muerte, y contar los fallecidos
reporte_tipo_muerte = fallecidos.groupby(['año', 'mes', 'criterio']).size().unstack(fill_value=0).reset_index()
```
